```python
import re
import pandas as pd
from selenium import webdriver
from webbrowser import Chrome
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time 
driver = webdriver.Chrome(ChromeDriverManager().install())
# driver=Chrome(Service(ChromeDriverManager().install()))
# driver = webdriver.Chrome(executablepath="C:\\Users\\kiran\\Downloads\\chromedriver_win32\\chromedriver.exe")
driver.get('https://www.imdb.com/?ref_=nv_home')
time.sleep(3)
driver.maximize_window()
#driver.find_element_by_id("iconContext-magnify").click()
driver.find_element(By.ID,"suggestion-search").send_keys("K.G.F: Chapter 2")
driver.find_element(By.ID,"suggestion-search-button").click()
driver.find_element(By.LINK_TEXT,'K.G.F: Chapter 2').click()
driver.find_element(By.LINK_TEXT,"User reviews").click()
# review = driver.find_element(By.XPATH,"//div[contains(@class,'imdb-u]ser-review')]")
#for i in range(1,70):
 #   driver.find_element(By.XPATH, "//button[@class='ipl-load-more__button']").click()


driver.find_element(By.XPATH, "//button[@class='ipl-load-more__button']").click()

    
all_ok = driver.find_elements(By.CLASS_NAME, "lister-item-content")
Rating = []
Title = []
Review = []
Name_Date = []
for i in all_ok:
    sen = i.text
    print(sen)
    sen = sen.split("\n")
    Rating.append(sen[0])
    Title.append(sen[1])
    Name_Date.append(sen[2])
    Review.append(sen[3])    
```

```python
import pandas as pd
kiran = pd.DataFrame({"name":Name_Date,"Title":Title,"comment":Review,"rating":Rating})
```

```python
kiran.to_csv("kiran.csv")
```
